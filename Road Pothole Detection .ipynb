```python
import cv2
import cvzone
import numpy as np
from ultralytics import YOLO
import os
path = r"D:\Computer Vision\road pothole detection  yolov8 custom segmentation\yolov8-roadpothole-detection-main"
os.chdir(path=path)



# Model 
model = YOLO("best.pt")

classes = model.names

cap = cv2.VideoCapture(r"p.mp4")

while True :
    ret , frame = cap.read()
    if ret == False :
        print("Error ,Loading  Frame")
        break
    
    
    frame = cv2.resize(frame , (1020,700))
    h , w , c = frame.shape
    
    results = model.predict(frame)
    
    for i in results :
        boxes = i.boxes
        masks = i.masks
    
    if masks :
        masks = masks.data.cpu()
        for seg , box in zip(masks.data.cpu().numpy() , boxes) :
            seg = cv2.resize(seg , (w,h))
            
            contours , _ = cv2.findContours((seg).astype(np.uint8) , cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_SIMPLE)
            
            for cont in contours :
                d = int(box.cls)
                c = classes[d]
                x1,y1,x2,y2 = cv2.boundingRect(cont)
                cv2.polylines(frame, [cont],True, color=(0, 0, 255), thickness=3)
                cvzone.putTextRect(frame, c, (x1,y1 - 10) , colorT=(0,255,255) , colorR=(0,0,0) , thickness=1 , scale=1.1)
                 
    cv2.imshow("Road Pothole Detection " , frame)
    
    if cv2.waitKey(1) == ord("q") :
        break
        
cv2.destroyAllWindows()
```
